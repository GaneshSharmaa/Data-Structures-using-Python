# Dictionary â€” Built-in Data Structure

- Dictionaries don't use indexes, instead they use key-value pairs
- Dictionaries are mutable in nature
- They are enclosed with curly brackets `{}`

### Defining a dictionary
Here, `1` is the key, and `"Python"` is the value
- `1` ---> `"Python"`
- `2` ---> `"C++"`
- `3` ---> `"Java"`

```python
dict_1 = {
    1: "Python",
    2: "C++",
    3: "Java",
    4: "JavaScript",
    5: "Swift"
}

print(dict_1)
print(type(dict_1))
```

### Accessing the elements of the dictionary

```python
print("Accessing the second value:", dict_1[2])     # 2 is not an index, instead it is key
# making clear with another example
```

```python
dict_2 = {
    "name": "Jack",
    "age": "22",
    "profession": "SDE",
    "location": "Mumbai",
    "email": "jack@company.com",
    "phone": "2318398445",
    "salary": "12 LPA"
}

print("Accesing the second value:", dict_2["age"])
```

### Changing the value of the dictionary

```python
dict_2["age"] = 23      # changing the age from 22 to 23
dict_2["location"] = "San Francisco"
print("After changing:", dict_2)
```

### Deleting the values in dictionary

Using `del` keyword

```python
del dict_2["phone"]
print("Dictionary after deletion:", dict_2)
```

Using `pop()` method - it also returns the deleted value

```python
deleted_pair = dict_2["email"]
print("Deleted pair:", deleted_pair)
print("Original dictionary:", dict_2)       # original dictionary remains intact
```

Using `popitem()` method - it returns the deleted key-value pair in tuple, and it takes no arguments

```python
dict_3 = {
    "brand": "BMW",
    "model": "X7",
    "type": "Luxury SUV",
    "launch date": "August 15, 2024",
    "engine type": "Diesel",
    "engine capacity": "2 liters"
}

print("Deleted key-value:", dict_3.popitem())
```

### Keys of the dictionary - returns the view object containing all the keys of the dictionary

```python
print("All the keys of the dictionary:", dict_3.keys())
```

### Values of the dictionary - returns the view object containing all the values of the dictionary

```python
print("All the values of the dictionary:", dict_3.values())
```

### Items of the dictionary - returns the view object containing all the items of the dictionary

```python
print("All the items of the dictionary:", dict_3.items())
```
